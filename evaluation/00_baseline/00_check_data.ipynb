```python
import pandas as pd
from _conf import BASELINE_ARCHIVES
from _utils import parse_solution_overview

baseline_data = pd.concat([parse_solution_overview(p) for p in BASELINE_ARCHIVES])
# Set non available sample sizes to None instead of 0 (which would harm the mean values)
baseline_data.loc[baseline_data["Path"].isna(), "SampleSize"] = None
baseline_data
```

```python
# Try parsing all instances
from _conf import BASELINE_ARCHIVES, INSTANCE_ARCHIVE
from _utils import get_instance, parse_solution_overview

instances_with_spaces = {}

for instance_name in baseline_data["Instance"].unique().tolist():
    instance = get_instance(instance_name, INSTANCE_ARCHIVE)
    if any(" " in f for f in instance.features):
        print(instance_name, "has a feature with spaces in it")
        print([f for f in instance.features if " " in f])
        instances_with_spaces[instance_name] = [
            f for f in instance.features if " " in f
        ]
print(instances_with_spaces)
```
