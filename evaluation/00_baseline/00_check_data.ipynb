```python
from _utils import parse_solution_overview
from _conf import BASELINE_ARCHIVES
import pandas as pd
baseline_data = pd.concat([parse_solution_overview(p) for p in BASELINE_ARCHIVES])
# Set non available sample sizes to None instead of 0 (which would harm the mean values)
baseline_data.loc[baseline_data["Path"].isna(), "SampleSize"]=None
baseline_data
```

```python
baseline_data.groupby(["Algorithm", "Settings", "Instance"]).size()
```

```python
# Make sure, we have the same amount of samples for each algorithm and instance
assert baseline_data.groupby(["Algorithm", "Settings", "Instance"]).size().min() == baseline_data.groupby(["Algorithm", "Settings", "Instance"]).size().max()
```

```python
# Make sure, every algorithm has run on the same number of instances
assert baseline_data.groupby(["Algorithm", "Settings"]).size().max() == baseline_data.groupby(["Algorithm", "Settings"]).size().min()
```
