```python
import tomllib
from pathlib import Path

with Path("../config.toml").open("rb") as f:
    config = tomllib.load(f)

config
```

```python
import pandas as pd

EXPERIMENT = "samplns_15min"
df = pd.read_json(config[EXPERIMENT]["iteration_info"])
df
```

```python
df.groupby("instance_name")["iteration"].count()
```

```python
df.loc[0, "iteration_info"]
```

```python
def time_of_last_update(iterations: list):
    end_lb = iterations[-1]["lb"]
    end_ub = iterations[-1]["ub"]
    for iteration in iterations:
        if iteration["lb"] == end_lb and iteration["ub"] == end_ub:
            return iteration["time"]
    return None
```

```python
df["time_of_last_update"] = df["iteration_info"].apply(time_of_last_update)
df
```

```python
assert (df["time_of_last_update"]<= df["runtime"]).all(), "Some time_of_last_update is greater than runtime, which cannot be"
```
