```python
import pandas as pd
from _conf import BASELINE_ARCHIVES, INSTANCE_ARCHIVE
from _utils import get_instance, parse_sample, parse_solution_overview

baseline_data = pd.concat(
    [parse_solution_overview(p) for p in BASELINE_ARCHIVES], ignore_index=True
)
# Set non available sample sizes to None instead of 0 (which would harm the mean values)
baseline_data.loc[baseline_data["Path"].isna(), "SampleSize"] = None
baseline_data
```

```python
from pathlib import Path
from _utils import parse_sample_file
from samplns.verify import have_equal_coverage

def fill_sample(sample, features):
    sample_ = []
    added_features = set()
    print(f"Sample has {len(sample)} configurations and {len(sample[0])} features")
    for conf in sample:
        for feature in features:
            if feature not in conf:
                conf[feature] = False
                added_features.add(feature)
        sample_.append(conf)
    print("Added features:", added_features)
    print(f"These are {len(added_features)} features out of {len(features)}")
    return sample_

for solution_path in Path("./results").rglob("*.csv"):
    instance_name = solution_path.parts[1]
    if "soletta" in instance_name:
        continue
    if "uclib" in instance_name:
        continue
    sample_idx = int(solution_path.name.split("_")[1].split(".")[0])
    print("Checking", instance_name, sample_idx)
    with open(solution_path) as f:
        sample = parse_sample_file(f)
    instance = get_instance(instance_name, INSTANCE_ARCHIVE)
    sample = fill_sample(sample, instance.features)
    print("Features:", instance.features)
    if not all(instance.is_feasible(conf, verbose=True) for conf in sample):
        print(sample[0])
        raise AssertionError()
    for feature in instance.features:
        if feature not in sample[0]:
            print("Feature missing:", sample[0])
            raise AssertionError()
    samples = baseline_data[baseline_data["Instance"] == instance_name]
    samples = samples[samples["Algorithm"] == "FIDE-YASA"]
    samples = samples[samples["Settings"]=="t2_m1_null"]
    samples = samples.dropna(subset=["Path"])  # drop all timed out samples
    if samples.empty or len(samples) == 1:
        print("Not enough samples to compare")
        continue
    baseline_sample = parse_sample(
        samples.iloc[0]["ArchivePath"], samples.iloc[0]["Path"]
    )
    baseline_sample = fill_sample(baseline_sample, instance.features)
    for feature in instance.features:
        if feature not in baseline_sample[0]:
            print(f"Feature '{feature}' missing in baseline:", baseline_sample[0])
            raise AssertionError()
    if not have_equal_coverage(instance, baseline_sample, sample):
        print(sample)
        print(samples.iloc[0])
        raise AssertionError()
    #print(instance_name, sample_idx, sample)
```
