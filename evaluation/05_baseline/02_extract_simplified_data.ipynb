```python
from _utils import parse_solution_overview

baseline_data = parse_solution_overview("./00_baseline_data.zip")
baseline_data = baseline_data.drop_duplicates(subset=["Instance", "AlgorithmID", "AlgorithmIteration", "ModelID", "Settings"])
baseline_data.loc[baseline_data["Path"].isna(), "SampleSize"] = None
baseline_data = baseline_data[["Instance", "#Variables", "#Clauses", "Algorithm", "Settings", "Time", "SampleSize", "Path"]]
baseline_data.rename(columns={"Time(s)": "runtime", "#Variables": "num_vars", "#Clauses": "num_clauses", "Instance": "instance_name", "Path": "path_to_baseline_sample", "Algorithm": "baseline_alg", "Settings": "baseline_alg_conf", "SampleSize": "sample_size"}, inplace=True)
baseline_data.to_json("./00_simple_baseline_data.json.zip")
baseline_data
```
