```python
import pandas as pd
from _conf import EXTENDED_BASELINE_DATA, LOWER_BOUNDS, PREPROCESSED_DATA

samplns_data = pd.read_json(PREPROCESSED_DATA)
samplns_data
```

```python
best_lower_bound = (
    samplns_data.groupby(["instance_name"])["lower_bound"]
    .max()
    .reset_index()
    .rename(columns={"lower_bound": "best_lower_bound"})
)
best_lower_bound.to_json(LOWER_BOUNDS)
best_lower_bound
best_lower_bound["internal_best_lower_bound"] = best_lower_bound["best_lower_bound"]
from _conf import EXTERNAL_BOUNDS

external_best_lb = (
    pd.concat([pd.read_csv(p) for p in EXTERNAL_BOUNDS])
    .groupby(["instance_name"])["best_lower_bound"]
    .max()
    .reset_index()
)
best_lower_bound = (
    pd.concat([best_lower_bound, external_best_lb])
    .groupby(["instance_name"])[["best_lower_bound", "internal_best_lower_bound"]]
    .max()
    .reset_index()
)
best_lower_bound
```

```python
best_upper_bound = (
    samplns_data.groupby(["instance_name"])["upper_bound"]
    .min()
    .reset_index()
    .rename(columns={"upper_bound": "best_upper_bound"})
)
best_upper_bound["internal_best_upper_bound"] = best_upper_bound["best_upper_bound"]
from _conf import EXTERNAL_BOUNDS

external_best_ub = (
    pd.concat([pd.read_csv(p) for p in EXTERNAL_BOUNDS])
    .groupby(["instance_name"])["best_upper_bound"]
    .min()
    .reset_index()
)
best_upper_bound = (
    pd.concat([best_upper_bound, external_best_ub])
    .groupby(["instance_name"])[["best_upper_bound", "internal_best_upper_bound"]]
    .min()
    .reset_index()
)
best_upper_bound
```

```python
# calculate the best known lower bound
# best_lower_bound = samplns_data.groupby(["instance_name"])["lower_bound"].max().reset_index().rename(columns={"lower_bound": "best_lower_bound"})
samplns_data = samplns_data.merge(best_lower_bound, on="instance_name")
```

```python
# calculate the best upper bound
# best_upper_bound = samplns_data.groupby(["instance_name"])["upper_bound"].min().reset_index().rename(columns={"upper_bound": "best_upper_bound"})
samplns_data = samplns_data.merge(best_upper_bound, on="instance_name")
```

```python
# calculate best baseline algorithm sample
baseline_data = pd.read_json(EXTENDED_BASELINE_DATA)
from _conf import EXTENDED_BASELINE_DATA

extended_baseline_data = pd.read_json(EXTENDED_BASELINE_DATA)
extended_baseline_data = pd.concat([baseline_data, extended_baseline_data])
baseline_data = extended_baseline_data
best_baseline = (
    baseline_data.groupby(["instance_name"])[["sample_size"]]
    .min()
    .reset_index()
    .rename(columns={"sample_size": "best_baseline_sample_size"})
)
samplns_data = samplns_data.merge(best_baseline, on="instance_name", how="left")
```

```python
performance_table = samplns_data.groupby("instance_name")[
    [
        "num_vars",
        "num_clauses",
        "best_baseline_sample_size",
        "baseline_sample_size",
        "upper_bound",
        "lower_bound",
        "best_upper_bound",
        "internal_best_upper_bound",
        "best_lower_bound",
        "internal_best_lower_bound",
    ]
].mean()
performance_table.sort_values(["num_vars", "num_clauses"], inplace=True)
performance_table["Improvement"] = 100 * (
    1
    - (
        performance_table["upper_bound"]
        / performance_table["best_baseline_sample_size"]
    )
)
performance_table["Max_Improvement"] = 100 * (
    1
    - (
        performance_table["best_upper_bound"]
        / performance_table["best_baseline_sample_size"]
    )
)
# checking consistency is important.
assert (
    performance_table["best_lower_bound"]
    <= performance_table["best_baseline_sample_size"]
).all()
assert (
    performance_table["best_upper_bound"] >= performance_table["best_lower_bound"]
).all()
performance_table
```

```python
# Code to query the time of the last change in lower or upper bound.


# Collect the data from the iteration_info events.
class EventCollector:
    def __init__(self):
        self.events = {
            "time": [],
            "val": [],
            "type": [],
            "instance": [],
            "path": [],
            # "alg": [],
            "initial_sample_size": [],
            "y": [],
            "final": [],
        }

    def __call__(self, row):
        it_data = row["iteration_info"]

        def add(val, lbub, time, final=False):
            if time > 900:
                return
            self.events["time"].append((time + row["time_used_by_yasa"]) / 60)
            self.events["val"].append(val)
            self.events["type"].append(lbub)
            self.events["instance"].append(row["instance_name"])
            # self.events["alg"].append(row["baseline_alg"])
            self.events["path"].append(row["path_to_baseline_sample"])
            self.events["initial_sample_size"].append(row["baseline_sample_size"])
            self.events["y"].append(100 * (val / row["best_lower_bound"]))
            self.events["final"].append(final)

        # add a zero entry
        add(0, "Lower", 0)
        add(row["baseline_sample_size"], "Upper", 0)
        for event in it_data:
            add(
                event["lb"],
                "Lower",
                event["time"],
                final=event["lb"] == row["lower_bound"],
            )
            add(
                event["ub"],
                "Upper",
                event["time"],
                final=event["ub"] == row["upper_bound"],
            )


ec = EventCollector()
samplns_data.apply(ec, axis=1)
t_events = pd.DataFrame(ec.events)
t_ = t_events.groupby(["instance", "path"])["time"].max().reset_index()


time_until_last_change = (
    t_events[t_events["final"]]
    .groupby(["instance", "path", "type"])[["time"]]
    .min()
    .reset_index()
    .groupby(["instance", "type"])[["time"]]
    .mean()
)

unsolved_symbol = "$*$"


def time_as_latex(time_s):
    secs = round(time_s)
    if secs == 0:
        return "$<\\SI{1}{\\second}$"
    if secs > 900:
        mins = round(secs / 60)
        return "\\SI{" + str(mins) + "}{\\minute}"
    return "\\SI{" + str(secs) + "}{\\second}"


from _conf import AVG_TIME_LONG_RUN


def get_time_info(instance):
    d = time_until_last_change.groupby("instance")["time"].max().to_dict()
    d_ = pd.read_csv(AVG_TIME_LONG_RUN).set_index("instance")["time"].to_dict()
    if instance not in d:
        return unsolved_symbol
    else:
        assert instance in d_
        return f"{time_as_latex(d[instance]*60)} ({time_as_latex(d_[instance])})"


time_until_last_change

performance_table = performance_table.merge(
    time_until_last_change.groupby("instance")["time"].max(),
    left_index=True,
    right_index=True,
)

performance_table["latex_time"] = performance_table.index.to_series().apply(
    get_time_info
)
# performance_table
```

```python
def to_latex_table(t: pd.DataFrame, extra=None) -> str:
    """
    Convert a pandas dataframe to a latex table.
    """
    tex_table = "\\begin{tabular}"
    tex_table += "{l " + " ".join(["r"] * (len(t.columns) - 1)) + "}\n"
    tex_table += " & ".join(t.columns) + "\\\\\n"
    tex_table += "\\hline\n"
    for idx in t.index:
        tex_table += f"{' & '.join(str(x) for x in t.loc[idx])}\\\\\n"
    if extra:
        tex_table += "\n"
        tex_table += extra + "\n"
        tex_table += "\n"

    tex_table += "\\end{tabular}"
    return tex_table


def shorten_name(name):
    threshold = 20
    if len(name) > threshold:
        return name[: threshold - 3] + "..."
    return name


def si_num(num):
    return "\\num[text-series-to-math=true]{" + str(int(num)) + "}"


def si_num_bold(num):
    return "\\textbf{" + si_num(num) + "}"


performance_table.sort_values(by=["num_vars", "num_clauses"], inplace=True)

data = {
    "Feature Model": [],
    "$|\\features|$": [],
    "$|\\clauses|$": [],
    "Best": [],
    "SampLNS": [],
    "LB": [],
    "Savings": [],
    "Time": [],
}
t = performance_table
num_opt_baseline = 0
num_potential_opt_baseline = 0
num_opt_samplns = 0
num_savings = 0

for idx in t.index:
    data["Feature Model"].append(shorten_name(idx.replace("_", "\\_")))
    data["$|\\features|$"].append(si_num(int(t["num_vars"][idx])))
    data["$|\\clauses|$"].append(si_num(int(t["num_clauses"][idx])))
    best_baseline = int(t["best_baseline_sample_size"][idx])
    best_lb = t["best_lower_bound"][idx]
    if best_baseline == best_lb:
        data["Best"].append("\\textbf{" + si_num(best_baseline) + "}")
        num_opt_baseline += 1
    else:
        data["Best"].append(si_num(best_baseline))
    if best_baseline <= t["best_upper_bound"][idx]:
        num_potential_opt_baseline += 1
    samplns = int(round(t["upper_bound"][idx]))
    if samplns == best_lb:
        num_opt_samplns += 1
        data["SampLNS"].append(
            "\\textbf{"
            + si_num(samplns)
            + "} (\\textbf{"
            + si_num(int(t["best_upper_bound"][idx]))
            + "})"
        )
        lb = int(round(t["lower_bound"][idx]))
        best_lb = int(t["best_lower_bound"][idx])
        data["LB"].append(
            (si_num_bold(lb) if lb == best_lb else si_num(lb))
            + " ("
            + si_num_bold(best_lb)
            + ")"
        )
    else:
        s = si_num(samplns) + " ("
        if t["best_upper_bound"][idx] == best_lb:
            s += "\\textbf{"
        else:
            s += "{"
        s += si_num(int(t["best_upper_bound"][idx])) + "})"
        data["SampLNS"].append(s)

        if t["best_upper_bound"][idx] == best_lb:
            data["LB"].append(
                si_num(int(round(t["lower_bound"][idx])))
                + " ("
                + si_num_bold(int(t["best_lower_bound"][idx]))
                + ")"
            )
        else:
            data["LB"].append(
                si_num(int(round(t["lower_bound"][idx])))
                + " ("
                + si_num(int(t["best_lower_bound"][idx]))
                + ")"
            )
    if t["Improvement"][idx] > 0:
        num_savings += 1
    data["Savings"].append(
        "\\SI{"
        + f"{round(t['Improvement'][idx], 2):.2g}"
        + "}{\\percent} (\\SI{"
        + f"{round(t['Max_Improvement'][idx], 2):.2g}"
        + "}{\\percent})"
    )
    data["Time"].append(t["latex_time"][idx])

percentage_opt_baseline = (
    "\\SI{" + str(round(100 * num_opt_baseline / len(t))) + "}{\\percent}"
)
percentage_opt_samplns = (
    "\\SI{" + str(round(100 * num_opt_samplns / len(t))) + "}{\\percent}"
)
percentage_improved = "\\SI{" + str(round(100 * num_savings / len(t))) + "}{\\percent}"

extra = (
    "\\hline \n \\ \\\\ \n \\textbf{optimality} & & & "
    + str(num_opt_baseline)
    + " & $\\geq "
    + str(num_opt_samplns)
    + "$ & & \\\\ & & & ["
    + percentage_opt_baseline
    + "] &["
    + percentage_opt_samplns
    + "] & & \\\\ \n \\textbf{improvements} & & & & &\\multicolumn{2}{r}{"
    + si_num(num_savings)
    + " ["
    + percentage_improved
    + "]} \\\\ \\ \\\\ \\ \\\\"
)

print(to_latex_table(pd.DataFrame(data=data), extra))
```
