## Simplify the data for later analysis

```python
from _utils import parse_solution_overview

data_initial = parse_solution_overview(INPUT_SAMPLE_ARCHIVE)
data_initial = data_initial.drop_duplicates(
    subset=["Instance", "AlgorithmID", "AlgorithmIteration", "ModelID", "Settings"]
)
data_initial.loc[data_initial["Path"].isna(), "SampleSize"] = None
data_initial = data_initial[
    [
        "Instance",
        "#Variables",
        "#Clauses",
        "Algorithm",
        "Settings",
        "Time(s)",
        "SampleSize",
        "Path",
    ]
]
data_initial.rename(
    columns={
        "Time(s)": "runtime",
        "#Variables": "num_vars",
        "#Clauses": "num_clauses",
        "Instance": "instance_name",
        "Path": "path_to_baseline_sample",
        "Algorithm": "baseline_alg",
        "Settings": "baseline_alg_conf",
        "SampleSize": "sample_size",
    },
    inplace=True,
)
data_initial.to_json("./00_simple_baseline_data.json.zip")
data_initial
```
